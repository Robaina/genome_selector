# Example using clustering to select subset

```python
import pandas as pd
from genomeselect.cluster_selector import select_representative_genomes_clustering


# Load your data
ec_path = "data/feature_prevalence.genomes.tsv"
abundance_path = "data/taxonomic_abundance.genomes.tsv"

ec_data = pd.read_csv(ec_path, index_col="id_genome", sep="\t")
abundance_data = pd.read_csv(abundance_path, index_col="id_genome", sep="\t")

# Run the selection algorithm
k = 5
selected_genomes = select_representative_genomes_clustering(
    ec_data, abundance_data, n_clusters=k
)

print("Selected representative genomes:")
print(selected_genomes)
```

```python
from genomeselect.cluster_selector import plot_ec_coverage_venn

plot_ec_coverage_venn(ec_data, selected_genomes, figsize=(8, 6))
```

```python
from genomeselect.cluster_selector import plot_ec_cumulative

plot_ec_cumulative(ec_data, abundance_data, figsize=(8, 6))
```
